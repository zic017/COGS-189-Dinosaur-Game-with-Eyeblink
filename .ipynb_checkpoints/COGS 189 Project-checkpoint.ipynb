```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("OpenBCI-RAW-2023-02-23_12-04-44.txt","r")
raw_data = file.readlines()
file.close()
```

```python
len(raw_data)
```

```python
raw_data
```

```python
data = [x.strip().split(",") for x in raw_data[5:500]]
data
```

```python
len(data)
```

```python
res = []
for arr in data:
    temp = [float(x.strip()) for x in arr[:-3]]
    res.append(temp)
res = np.array(res)
res
```

```python
plt.plot(res)
```

```python
# flip axes?
f, ax = plt.subplots()
ax.plot(res)

newx = ax.lines[0].get_ydata()
newy = ax.lines[0].get_xdata()

ax.lines[0].set_xdata(newx)
ax.lines[0].set_ydata(newy)
```
