```python
import math
def divide_dataset(data, time):
    epoch = 500
    ran = math.ceil(len(time) / epoch)
    data_subsets = []
    time_subsets = []
    start = 0
    for i in range(ran):
        if start + epoch >= len(time):
            data_subsets.append(data.iloc[start:,:])
            time_subsets.append(time[start:])       
        else:
            data_subsets.append(data.iloc[start:start+epoch,:])
            time_subsets.append(time[start:start + epoch])
            start = start + epoch
    return data_subsets, time_subsets
```

```python
def find_max_and_min_avg(arr):
    max_sum_1 = 0
    min_sum_1 = 0
    max_sum_2 = 0
    min_sum_2 = 0
    for channel in arr:
        max_sum_1 += max(channel['fp1'].tolist())
        min_sum_1 += min(channel['fp1'].tolist())
        max_sum_2 += max(channel['fp2'].tolist())
        min_sum_2 += min(channel['fp2'].tolist())
    avg_max_1 = max_sum_1 / len(arr)
    avg_min_1 = min_sum_1 / len(arr)
    avg_max_2 = max_sum_2 / len(arr)
    avg_min_2 = min_sum_2 / len(arr)
    return avg_max_1, avg_min_1, avg_max_2, avg_min_2
        
```
